```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("student_score.csv")
print(df.head())
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#Drop unnamed colunm
```

```python
df = df.drop("Unnamed: 0", axis =1)
print(df.head())
```

```python
#change WklyStudyHours colunms
```

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("05-Oct" , "5-10")
df.head()
```

```python
# gender distribution
```

```python
plt.figure(figsize=(4 , 4))
ax = sns.countplot(data = df , x = "Gender" )
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
#from above data we have analysed the number of femles is more than number of males
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean' , "WritingScore":'mean'})
print(gb)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(5, 5))
sns.heatmap(gb, annot=True)
plt.title("Relation between parents education and students score")
plt.show()
```

```python
#From above analysis we have concluded that education of parents have impact on stuents score
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean' , "WritingScore":'mean'})
print(gb1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(5, 5))
sns.heatmap(gb1, annot=True)
plt.title("Relation between parents  MaritalStatus and students score")
plt.show()
```
